```python
from stixdcpy.quicklook import LightCurves
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../eui_hri_list.csv')
```

```python
%matplotlib inline
for i in range(len(df)):
    start_time = df['START DATE & TIME'][i][:11] + ':' + df['START DATE & TIME'][i][11:13] + ':' + df['START DATE & TIME'][i][13:15] + '.' +  df['START DATE & TIME'][i][15:]  
    end_time = df['END DATE & TIME'][i][:11] + ':' + df['END DATE & TIME'][i][11:13] + ':' + df['END DATE & TIME'][i][13:15] + '.' +  df['END DATE & TIME'][i][15:]  
    lc = LightCurves.from_sdc(start_utc=start_time, end_utc=end_time, ltc=False)
    lc.peek()
    plt.title(f"[{i}] {df['SOOPNAME'][i]}")
    plt.savefig(f"stx_quicklooks/stx_quicklook_{i}_{df['SOOPNAME'][i]}.png", dpi=300)
    plt.show()
```
