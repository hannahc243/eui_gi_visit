```python
from stixdcpy.quicklook import LightCurves
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../eui_hri_list.csv')
```

```python
%matplotlib inline
for i in range(len(df)):
    start_time = pd.to_datetime(df['START DATE & TIME'][i][:11] + ':' + df['START DATE & TIME'][i][11:13] + ':' + df['START DATE & TIME'][i][13:15] + '.' +  df['START DATE & TIME'][i][15:])  
    end_time = pd.to_datetime(df['END DATE & TIME'][i][:11] + ':' + df['END DATE & TIME'][i][11:13] + ':' + df['END DATE & TIME'][i][13:15] + '.' +  df['END DATE & TIME'][i][15:] ) 
    lc = LightCurves.from_sdc(start_utc=start_time, end_utc=end_time, ltc=False)
    fig,ax=plt.subplots()
    lc.peek(ax=ax)
    ax.set_xlim(start_time,end_time)
    plt.title(f"[{i}] {df['SOOPNAME'][i]}")
    #plt.savefig(f"../stx_quicklooks/stx_quicklook_{i}_{df['SOOPNAME'][i]}.png", dpi=300)
    plt.show()

```
